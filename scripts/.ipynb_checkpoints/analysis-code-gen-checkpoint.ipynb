```python
from sklearn.metrics import r2_score

def calc_r2(observed,predicted):
    
    global lnconvert
    
    # Replace Zero values
    #
    tmp = []
    omid = np.median(np.array(observed))
    if (omid == 0.0):
        omid = np.mean(np.array(observed))
    for index in range(0,len(observed)):
        if (observed[index] == 0.0):
            tmp.append(omid)
        else: 
            tmp.append(observed[index])
            
    o = np.array(tmp)
    
    # Replace formula predicted negative values
    #
    tmp = []
    ptmp = np.array(predicted)
    #pmid = np.median(np.array(predicted))
    pmid=predicted[np.where(ptmp > 0.0, ptmp, np.inf).argmin()]
    print(pmid)
    
    #if (pmid <= 0.0):
    #    pmid = np.mean(np.array(predicted))
    
    for index in range(0,len(predicted)):
        if (predicted[index] <= 0.0):
            tmp.append(pmid)
        else: 
            tmp.append(predicted[index])   
    
    p = np.array(tmp)
    
    #print("-------------------")
    #print(o)
    #print("-------------------")
    #print(p)
    
    if (lnconvert == True):
        o = np.log(o)
        p = np.log(p)
    
    #print("-------------------")
    #print(o)
    #print("-------------------")
    #print(p)
    
    r2 = r2_score(o,p)
    print("*********************************************")
    print('R^2: ', r2) 
    print("*********************************************")
    
    #x =range(0,len(o))
    #
    #name = "ln compare " + str(r2)
    #plt.figure(name , figsize=(10, 10))
    #plt.title(name)
    #plt.xlabel('categories')
    #plt.ylabel('probability index')
    #plt.plot(x,o,'-r')
    #plt.plot(x,p,'-g')
    #plt.show
    
    return(r2)
```

```python
def find_order(data):
    
    atmp = []
    for i in range(0,len(data)):
        atmp.append( (i,data[i]) )
        
    #print("HERE")
    #print(atmp)
    
    atmp2 = sorted(atmp, key = lambda x: x[1])

    #print("THERE")
    #print(atmp2)
    
    l = []
    for i in range(0,len(atmp2)):
        l.append(atmp2[i][0])
        
    return(l)
```

```python
def plot_avg(name,data):
    global swivel
    global rsq_avg
    global year
    
    # Creat plot data
    x = data[0]
    y = data[1]
    count = 0
    
    # Find best swivel order
    print("Best swivel order for combined data")
    print(find_order(y))
    
    # Curve fit
    r = []
    try:
        popt, pcov = curve_fit(func, x, y)
    except:
        popt = [0,0,0]
    
    popt
    
    for i in range(0,len(x)):
        r.append(func(i, *popt))
        
    rsq = calc_r2(y,r)
    
    if (year == 8):
        rsq_avg = popt
        print("############################################### RSQ_AVG ################################################")
        print(rsq_avg)
        
    # Plot the data
    title = name+" combined avg "
    
    plt_title = title + ", r2=" + str(rsq) + " " + str(popt) 

    plt.figure(title, figsize=(10, 10))
    plt.title(plt_title)
    plt.xlabel('categories')
    plt.ylabel('probability index')
    
    max_value = max(y)
    #plt.ylim(0,12)
    #plt.ylim(0,max_value*1.2)
    
    if (swivel == 1):
        job_code = soc_swivel_code
    else:
        job_code = soc_new_code
    
    
    total_len = len(job_code) - 1
    
    # Plot points
    for index in range(0,total_len):  
        plabel = str(index) + " " + str(job_code[index+1][1])
        plt.plot(x[index],y[index],'o',label=plabel, markersize=4, color='black')
        #plt.plot(x[index], func(index, *popt) ,'or', markersize=2)
        plt.text(x[index],y[index],str(index))
        #print(x[index], func(index, *popt))
        
    # Plot line regression
    plt.plot(x, r ,'-g')

    #plt.legend(loc="upper left", bbox_to_anchor=(0.72, 0.3), ncol= 1, fontsize="small")
    plt.legend(loc="upper left", ncol= 1, fontsize="small")
    
    save_title = "../PLOTS/" + desired_file_output + title + ".png"
    save_title = save_title.replace(",","")
    save_title = save_title.replace(" ","_")
    plt.savefig(save_title, dpi=100, bbox_inches='tight')
    
    plt.show()
    plt.close()
```

```python
def color_corr_green(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = 'green' if val > 0.8 else 'black'
    return 'color: %s' % color
```

```python
def avg_data(avg_list): #list_to_avg):
    
    length = len(global_y_soc1[0][1])
    
    # Create plot data
    x = np.zeros(length)
    count = 0.0

    print(global_y_soc1)
    
    # Init to first found array ( (5 + 5) / 2 = 10 , no need to worry about first sample repeat )
    y = np.zeros(length)
    y = y + global_y_soc1[0][1]
    
    for to_avg in avg_list:
        for index in range(0,len(global_y_soc1)):
            #print(index,to_avg)
            cat = str(global_y_soc1[index][0])
            #print(cat,to_avg)
            if (cat.find(to_avg)>=0):
                #print(global_y_soc1[index][0])
                #print(global_y_soc1[index][1])
                y = (y + global_y_soc1[index][1]) / 2
    
    for index in range(0,23):
        x[index] = count
        count=count+1.0
        
    data = np.array([x,y])

    return(data)
```
